```python
import pandas as pd
df = pd.DataFrame({
    "loan": [1,1,1,2,2],
    "mob": [0,1,2,0,1],
    "state": ["DPD0","DPD1+","DPD30+","DPD0","DPD0"]
})
g = df.groupby("loan", group_keys=False)
df["mob_next"] = g["mob"].shift(-1)
df["state_t1"] = g["state"].shift(-1)
df["state_t"]  = df["state"]   # lỗi tiềm ẩn
```
